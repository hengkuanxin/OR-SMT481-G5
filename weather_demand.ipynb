```python
import zipfile
import pandas as pd

# Unzipping the file
with zipfile.ZipFile('2023-citibike-tripdata/202309-citibike-tripdata.zip', 'r') as zip_ref:
    zip_ref.extractall('extracted_files')

# Loading data into a DataFrame
df = pd.read_csv('extracted_files/202309-citibike-tripdata_2.csv')

# Display the DataFrame
df.head()
```

```python
import glob, os

path = '2023Autumn' # use your path
all_files = glob.glob(os.path.join(path , "*.csv"))

li = []

for filename in all_files:
    print(filename)
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

print("Added all csv dfs into a list")
```

```python
# Concatenate all DataFrames together into 1 mainframe
df = pd.concat(li, axis=0, ignore_index=True)
df.head()
```

```python
# Sort the DataFrame by the 'started_at' column -- loading could have been out of order
citibike_df = df
citibike_df.sort_values(by = 'started_at', inplace = True)
citibike_df.reset_index(drop = True, inplace = True) # reset the index
citibike_df.head()
```

```python
# Adjusting datetime

citibike_df['started_at'] = pd.to_datetime(citibike_df['started_at'], format='mixed')

# Localize to Eastern Time (America/New_York), which will respect DST (EDT and EST mix)
citibike_df['started_at'] = citibike_df['started_at'].dt.tz_localize("America/New_York", ambiguous=True, nonexistent='shift_forward')

citibike_df.head()
```

```python
# Load the NYC_Weather data into a DataFrame
weather_df = pd.read_csv('nyc-2023-weather-data.csv')

# Display the DataFrame
weather_df.head()
```

```python
# Convert 'time' column to datetime
weather_df['time'] = pd.to_datetime(weather_df['time'])

# Localize to Eastern Time (America/New_York), which automatically adjusts for DST
weather_df['time'] = weather_df['time'].dt.tz_localize("America/New_York", ambiguous=True, nonexistent='shift_forward')

weather_df.head()
```

```python
# Checking final dataframes
citibike_df.head()
```

```python
print(weather_df['time'].head(10))

# Check for non-datetime values using `apply` and `type`
non_datetime_values = weather_df[~weather_df['time'].apply(lambda x: isinstance(x, pd.Timestamp))]
print(non_datetime_values)
```

```python
weather_df.head()
```

```python
# Define Manhattan's geographical boundaries (approximate latitudes and longitudes)
min_lat, max_lat = 40.4774, 40.9176   # Latitude range for Manhattan
min_lon, max_lon = -74.0210, -73.9295  # Longitude range for Manhattan

# Filter the dataset based on latitude and longitude for Manhattan
manhattan_df = citibike_df[(citibike_df['start_lat'] >= min_lat) & 
                            (citibike_df['start_lat'] <= max_lat) & 
                            (citibike_df['end_lng'] >= min_lon) & 
                            (citibike_df['end_lng'] <= max_lon)]

# Check the filtered dataframe
manhattan_df.head()
```

```python
manhattan_df['hour'] = manhattan_df['started_at'].dt.hour
manhattan_df['date'] = manhattan_df['started_at'].dt.date
manhattan_df.head()
```

```python
manhattan_df['started_at'] = pd.to_datetime(manhattan_df['started_at'])

# Now you can extract the hour and date
manhattan_df['hour'] = manhattan_df['started_at'].dt.hour
manhattan_df['date'] = manhattan_df['started_at'].dt.date

# Check the dataframe
manhattan_df.head()
```

```python
# Aggregating ride count data
manhattan_df['ride_count'] = manhattan_df.groupby(['date', 'hour'])['ride_id'].transform('count')
manhattan_df.head()
```

```python

weather_df['hour'] = weather_df['time'].dt.hour
weather_df['date'] = weather_df['time'].dt.date
weather_df.head()
```

```python
import seaborn as sns

import matplotlib.pyplot as plt

# Merge the dataframes on the date column
merged_df = pd.merge(manhattan_df, weather_df, on=['date', 'hour'], how='inner')
merged_df.columns
```

```python
merged_df.head()
```

```python
# Preprocessing merged_df to remove uncomparable columns
corr_df = merged_df[['date', 'hour', 'ride_id', 'start_lat', 'start_lng', 'end_lat', 'end_lng', 'ride_count', 'temperature_2m (°C)', 'precipitation (mm)', 'rain (mm)', 'cloudcover (%)',
       'windspeed_10m (km/h)', 'winddirection_10m (°)']]
corr_df.head()
```

```python
# Define the weather conditions based on the criteria
def categorize_weather(row):
    if row['rain (mm)'] > 0:
        return 'Rainy'
    elif row['cloudcover (%)'] > 50:
        return 'Cloudy'
    elif row['precipitation (mm)'] == 0 and row['cloudcover (%)'] <= 50:
        return 'Sunny'
    else:
        return 'Normal'

# Apply the function to categorize each row's weather
corr_df['weather_condition'] = corr_df.apply(categorize_weather, axis=1)

# Check the new dataframe with weather conditions
corr_df.head()
```

```python
# Group by the weather condition and calculate the average ride_count
weather_demand = corr_df.groupby('weather_condition')['ride_count'].mean().reset_index()

# Check the demand across weather conditions
print(weather_demand)
```

```python
import matplotlib.pyplot as plt

# Plotting the demand by weather condition
plt.figure(figsize=(8, 6))
plt.bar(weather_demand['weather_condition'], weather_demand['ride_count'], color=['skyblue', 'lightgrey', 'orange', 'lightgreen'])
plt.title('Average Ride Count by Weather Condition')
plt.xlabel('Weather Condition')
plt.ylabel('Average Ride Count')
plt.show()
```

```python
# Convert 'date' column to datetime if it's not already
corr_df['date'] = pd.to_datetime(corr_df['date'], errors='coerce')

# Add a 'week' column based on the 'date' column
corr_df['week'] = corr_df['date'].dt.to_period('W')
```

```python
corr_df.head()
```

```python
# Group by the 'week' and 'weather' columns, then calculate the mean ride_count for each group
avg_ride_count_per_week_weather = corr_df.groupby(['week', 'weather_condition'])['ride_count'].mean().reset_index()

# Display the result
print(avg_ride_count_per_week_weather)
```

```python
# Save the DataFrame as a CSV file
avg_ride_count_per_week_weather.to_csv('avg_ride_count_per_week_weather.csv', index=False)

# Verify by printing the path
print("CSV file saved as 'avg_ride_count_per_week_weather.csv'")
```
